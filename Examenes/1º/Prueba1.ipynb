```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.Symbol("x")
```

#### Ejercicio 1

```python
def f(x):
    y = np.cos(5*x) - np.sin(np.sqrt(x**2 + 1))
    return y


N = np.array([f(-np.pi), f(-1) , f(0) , (1) , f(np.pi)]) # Creamos un array en el que cada elemento es el valor de la función en los puntos que queremos

print(N)
```

##### Ejercicio 2


```python
x_vals = np.linspace(-np.pi, np.pi, 80)

plt.plot(x_vals, f(x_vals)) # Representamos f en función de los valores de x entre -pi y pi con 80 valores
plt.title("Representación de f(x)")
plt.ylabel("y")
plt.xlabel("x")

plt.show()
```

#### Ejercicio 3

```python
g = ((x**3 - 8) / (x - 2)) + 10

g_s = sp.simplify(g) # Simplificamos

display(g_s) 
```
